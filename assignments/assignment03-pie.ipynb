```python
import pandas as pd

people = pd.read_csv('people-1000.csv')

people.head()
```

#### Pie Chart Male Vs. Female 
Refs:

```python
import pandas as pd
import matplotlib.pyplot as plt


# Count the occurrences of each category in the 'sex' column
sex_counts = people['Sex'].value_counts()

# Plot the pie chart
plt.figure(figsize=(8, 8))
plt.pie(sex_counts, labels=sex_counts.index, autopct='%1.1f%%', startangle=90, textprops={'fontsize': 16})

# Add a title
plt.title('Distribution of Sex in the Dataset', fontsize=16)

# Save the plot
plt.savefig('sex_pie_chart.png')

# Display the pie chart
plt.show()
```
