```python
import pandas as pd

data = pd.read_csv('weatherreadings1.csv')

data.head()
```

```python
# Change date format

# Convert the 'reportStartDateTime' column to datetime
data['reportStartDateTime'] = pd.to_datetime(data['reportStartDateTime'])

# Inspect the data to confirm the conversion
print(data[['reportStartDateTime']].tail())  # View the first few rows of the 'time' column
```

```python
import matplotlib.pyplot as plt

# Plot the data
plt.figure(figsize=(10, 6))
plt.plot(data['reportStartDateTime'], data['dryBulbTemperature_Celsius'], label='Temperature Over Time', linewidth=2)

# Add labels and title
plt.title('Temperature Over Time', fontsize=16)
plt.xlabel('Time', fontsize=14)
plt.ylabel('Temperature (Â°C)', fontsize=14)
plt.xticks(rotation=45)
plt.grid(alpha=0.3)
plt.legend()
    
# Save the plot
plt.savefig('temperature_plot.png')

# Show the plot
plt.show()
```
